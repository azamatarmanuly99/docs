```python
import os

def process_pdb(input_pdb_path, output_pdb_path):
    chain_h = []
    chain_l = []
    chain_b = []

    with open(input_pdb_path, 'r') as pdb_file:
        lines = pdb_file.readlines()

    residue_number_l = 1
    residue_number_h = 0
    last_residue_seen_h = None
    last_residue_seen_l = None
    for line in lines:
        if line.startswith("ATOM") or line.startswith("HETATM"):
            chain_id = line[21]
            residue_number = int(line[22:26])

            if chain_id == 'A':
                if residue_number <= 114:
                    if last_residue_seen_h != residue_number:
                        residue_number_h += 1
                        last_residue_seen_h = residue_number
                    new_line = line[:21] + 'H' + f"{residue_number_h:4d}" + line[26:]
                    chain_h.append(new_line)
                else:
                    if last_residue_seen_l != residue_number:
                        if residue_number_l <= 107:
                            last_residue_seen_l = residue_number
                            residue_number_l += 1
                        else:
                            continue
                    new_line = line[:21] + 'L' + f"{residue_number_l:4d}" + line[26:]
                    chain_l.append(new_line)
            elif chain_id == 'B':
                chain_b.append(line)

    with open(output_pdb_path, 'w') as out_pdb:
        for line in chain_h + chain_l + chain_b:
            out_pdb.write(line)
        out_pdb.write("TER\nEND\n")

input_folder_path = "/content/input13"
output_folder_path = "/content/output14"
os.makedirs(output_folder_path, exist_ok=True)

for i in range(30):  # Iterate through designs
    for j in range(10):  # Iterate through cycles for each design
        input_file_name = f"design_{i}_dldesign_0_cycle{j}.pdb"
        input_pdb_path = os.path.join(input_folder_path, input_file_name)
        output_file_name = f"design_{i}_dldesign_0_cycle{j}_output.pdb"
        output_pdb_path = os.path.join(output_folder_path, output_file_name)

        if os.path.exists(input_pdb_path):
            process_pdb(input_pdb_path, output_pdb_path)
        else:
            print(f"File not found: {input_pdb_path}")
# Call the function to process the PDB file
process_pdb(input_pdb_path, output_pdb_path)
```
